```python
import requests
from bs4 import BeautifulSoup
import csv

# Base URL
base_url = 'https://quotes.toscrape.com'
headers = {'User-Agent': 'Mozilla/5.0'}

# List to store filtered quotes
quotes = []

# Function to scrape one page
def scrape_page(soup, quotes):
    for quote in soup.find_all('div', class_='quote'):
        text = quote.find('span', class_='text').text
        author = quote.find('small', class_='author').text
        tags = ', '.join(tag.text for tag in quote.find_all('a', class_='tag'))

        # Filter only Albert Einstein quotes
        if "albert einstein" in author.lower():
            quotes.append({
                'Text': text,
                'Author': author,
                'Tags': tags
            })

# Function to scrape all pages
def scrape_all_pages(url):
    while url:
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.text, 'html.parser')

        scrape_page(soup, quotes)

        next_page = soup.find('li', class_='next')
        url = base_url + next_page.find('a')['href'] if next_page else None

# Call function
scrape_all_pages(base_url)

# Check how many quotes found
print("Total Einstein quotes found:", len(quotes))

# Save to CSV
with open('quotes_Einstein.csv', 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=['Text', 'Author', 'Tags'])
    writer.writeheader()
    writer.writerows(quotes)

print("Einstein quotes saved successfully.")
```
