```python
# 1. Load dataset
import pandas as pd

df = pd.read_csv("Processed_Reviews.csv")
df.head()

# Check column names
print(df.columns)

# Remove spaces in column names
df.columns = df.columns.str.strip()


# 2. Check missing & duplicates
df.isnull().sum()
df.duplicated().sum()
df = df.drop_duplicates()


# 3. Prepare data (AUTO DETECT TEXT COLUMN 💡)

# assume last column = label, first column = text
X = df.iloc[:, 0]   # text column (auto ambil first column)
y = df.iloc[:, -1]  # label column (auto ambil last column)


# 4. Vectorization (TF-IDF)
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vect = TfidfVectorizer()
X = tfidf_vect.fit_transform(X)


# 5. Split data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)


# 6. Naïve Bayes model
from sklearn.naive_bayes import MultinomialNB

nb_clf = MultinomialNB().fit(X_train, y_train)


# 7. Prediction
nb_predicted = nb_clf.predict(X_test)


# 8. Evaluation
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print("Naive Bayes Classifier Report:")
print(classification_report(y_test, nb_predicted))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, nb_predicted))

print("\nNaive Bayes Error Rate:", 1 - accuracy_score(y_test, nb_predicted))
```
